```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
documents = nltk.sent_tokenize(paragraph)
```

```python
import re
corpus = []
for i in documents:
    review = re.sub("[^a-zA-Z]"," ",i)
    review = review.lower()
    corpus.append(review)
```

```python
corpus[0]
```

```python
print(stopwords.words('english'))
```

```python
stemmer = PorterStemmer()
```

```python
cleaned_corpus = []
for s in corpus:
    words = nltk.word_tokenize(s)
    clean_words = [stemmer.stem(word) for word in words if word not in stopwords.words('english')]
    cleaned_corpus.append(" ".join(clean_words))
```

```python
cleaned_corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(binary=False,ngram_range=(3,3))
```

```python
X = cv.fit_transform(corpus)
```
